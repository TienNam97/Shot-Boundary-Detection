```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
def Calculator(file_save_path, file_gt_path):
    # C: correct detection
    # M: missed detection
    # F: false detection
    C, M, F = 0, 0, 0
    result = np.loadtxt(file_save_path)[:, 1]
    gt =  np.loadtxt(file_gt_path)[:, 1]
    
    for x in result:
        if np.abs(x - find_nearest(gt,x)) <= 5:
            C += 1
    
    M = len(gt) - C
    F = len(result) - C
    Recall = C / (C + M)
    Precision = C / (C + F) if (C + F) != 0 else 0
    return Recall, Precision
```

#Độ chính xác phương pháp dựa vào Histogram trên tập RAI 

```python
Recall, Precision, i, R, P = 0, 0, 0, 0, 0
for file in os.listdir("./RaiSceneDetection/RAIDataset"):
    if (file.endswith(".MP4")|file.endswith(".mp4")):
        i += 1
        videos_src_path = os.path.join("./RaiSceneDetection/RAIDataset", file)
        file_name = Path(videos_src_path).stem
        save_file = "".join([file_name,"_result.txt"])
        gt_file = "".join([file_name,"_gt.txt"])
        file_save_path = os.path.join("./RaiSceneDetection/Histogram", save_file)
        file_gt_path = os.path.join("./RaiSceneDetection/RAIDataset", gt_file)
        R, P = Calculator(file_save_path, file_gt_path)
        Recall += R
        Precision += P

Recall_Rai_Hist = Recall/i
Precision_Rai_Hist = Precision/i

print("Recall: ",Recall_Rai_Hist)
print("Precision: ",Precision_Rai_Hist)
```
